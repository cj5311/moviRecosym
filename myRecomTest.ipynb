## 하이브리드(Hybrid)

* 컨텐츠 기반 필터링과 협업 필터링을 조합한 방식
* 많은 하이브리드 방식이 존재
* 실습에서는 협업 필터링으로 임베딩을 학습하고 컨텐츠 기반 필터링으로 유사도 기반 추천을 수행하는 추천 엔진 개발

```python
import numpy as np
from sklearn.decomposition import randomized_svd, non_negative_factorization
from surprise import Dataset
import myRecom as mr
```

```python
#데이터 로드드
data  = Dataset.load_builtin('ml-100k',prompt=False)
raw_data = np.array(data.raw_ratings, dtype=int)
```

```python
#인접행렬생성
adj_matrix = mr.adjacentRow(raw_data, type = 'rating')
```

#### 임베딩(컨텐츠기반) --> 코사인유사도(협업기반) --> 사용자기반, 아이템기반 추천

```python
#사용자 기반 추천 (임베딩 : SVD, 유사도 : 코사인)
#나와 비슷한 취향을 가진 다른 사용자의 행동을 추천
#사용자 특징 벡터의 유사도 사용

_,_,_ = mr.hybridRecommend(adj_matrix, i=0 , type = 'user')
```

```python
#사용자 기반 추천 (임베딩 : SVD, 유사도 : 코사인)
#나와 비슷한 취향을 가진 다른 사용자의 행동을 추천
#사용자 특징 벡터의 유사도 사용

_,_,_ = mr.hybridRecommend (adj_matrix, i=0, embediing ='svd',similarity ='cos',type = 'item')
```

#### 비음수 행렬 분해를 사용한 하이브리드 추천

```python
_,_,_ = mr.hybridRecommend (adj_matrix, i=0, embediing ='nmf',similarity ='cos',type = 'user')
```
